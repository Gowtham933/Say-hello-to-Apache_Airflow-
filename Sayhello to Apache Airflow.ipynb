```python
import os

# Ensure the directory exists before writing the file
os.makedirs("/content/airflow/dags", exist_ok=True)
print("My_directory exist")

dag_code = """
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime

def print_hello():
    print("Hello, Airflow!")

default_args = {"start_date": datetime(2024, 2, 7)}

with DAG("hello_airflow", schedule_interval="@daily", default_args=default_args, catchup=False) as dag:
    task = PythonOperator(
        task_id="say_hello",
        python_callable=print_hello
    )

    task
"""

# Use a single slash for the file path
with open("/content/airflow/dags/hello_airflow.py", "w") as f:
    f.write(dag_code)
```

```python
!airflow dags list
!airflow dags unpause hello_airflow  # Unpause DAG
!airflow dags trigger hello_airflow  # Run DAG
```

```python
!airflow tasks list hello_airflow
!airflow tasks test hello_airflow say_hello 2024-02-07
```
